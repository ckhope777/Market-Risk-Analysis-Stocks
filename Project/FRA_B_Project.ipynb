# Stock Price Analysis

```python
numeric_columns = df.select_dtypes(include='number').columns 

for i, stock in enumerate (numeric_columns):
    plt.figure(figsize=(10, 6))
    plt.scatter(data=df, x='Date', y=stock)  
    plt.title(f'{stock} Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.grid(True)
    plt.show()
```

# Returns and Volatility Analysis


## Return Calculation

```python
Return_of_Stocks =  np.log(df.drop(['Date'],axis=1)).diff(axis = 0)

pd.set_option('display.max_rows', None) 

Return_of_Stocks
```

**Inference**
-


## Average Returns

```python
StockMeans = Return_of_Stocks.mean()  
StockMeans.sort_values()
```

**Inference**
-


## Volatility

```python
StockStdDev = Return_of_Stocks.std()  
print(StockStdDev.sort_values())
```

**Inference**
-


## Visualizing Returns and Volatility

```python
data = pd.DataFrame({'Means': StockMeans, 'StdDev': StockStdDev})  
```

```python
import seaborn as sns

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
plt.scatter(data=data, x='StdDev', y='Means')  
plt.axhline(y=0,linestyle='--', color = "red")

for index, row in data.iterrows():
    ax.text(row['StdDev'], row['Means'], index)
plt.xlabel('volatility')
plt.ylabel('Net returns')
plt.title('Net return vs Volatility for stocks')
plt.grid()
plt.show()
```
